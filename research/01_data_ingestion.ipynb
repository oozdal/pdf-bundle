```python
import langchain
from langchain.document_loaders import PyPDFDirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
## Lets Read the document
def read_doc(directory):
    file_loader=PyPDFDirectoryLoader(directory)
    documents=file_loader.load()
    return documents
```

```python
doc=read_doc('documents/')
len(doc)
```

```python
def chunk_data(docs,chunk_size=800,chunk_overlap=50):
    text_splitter=RecursiveCharacterTextSplitter(chunk_size=chunk_size,chunk_overlap=chunk_overlap)
    doc=text_splitter.split_documents(docs)
    return doc
```

```python
documents=chunk_data(docs=doc)
len(documents)
```

```python
sentences = [document.page_content for document in documents]
```

```python
import pinecone

# Connect to Pinecone
#pinecone.init(
#    api_key="00de6be5-b2b9-49bc-8630-a2c2ed8d94a6",
#    environment="us-west-2"
#)
#index_name="langchainvector"
pinecone.create_index("sentence_index", dimension=300)

def embed_sentence(sentence):
    # Placeholder function to generate embeddings for a sentence (replace with your own method)
    # This function should return a 300-dimensional vector representing the embedding of the input sentence
    # For demonstration purposes, let's just use a random vector
    import numpy as np
    return np.random.rand(300)

def store_sentence(sentence):
    # Embed the sentence
    embedding = embed_sentence(sentence)
    # Store the sentence and its embedding in Pinecone
    pinecone.upsert_items("sentence_index", [sentence], [embedding])

def find_similar_sentences(query_sentence, k=5):
    # Embed the query sentence
    query_embedding = embed_sentence(query_sentence)
    # Search for similar sentences using cosine similarity
    results = pinecone.query("sentence_index", query_embedding, top_k=k)
    return results

# Example usage
# Store some example sentences
sentences_to_store = [
    "The quick brown fox jumps over the lazy dog.",
    "A stitch in time saves nine.",
    "Actions speak louder than words.",
    "Beauty is in the eye of the beholder.",
    "A penny saved is a penny earned."
]

for sentence in sentences_to_store:
    store_sentence(sentence)

# Find similar sentences to a query
query_sentence = "A fox is jumping over a dog."
similar_sentences = find_similar_sentences(query_sentence)
print("Similar sentences to '{}':".format(query_sentence))
for i, (sentence, _) in enumerate(similar_sentences):
    print("{}. {}".format(i+1, sentence))
```
